## Using mean of nearest distance in every dim


```python
l=np.zeros((100,416))
```

```python
l2= np.zeros((13,32))
l3 = np.zeros((13,32))
```

```python
for p in range(13):
  
  for k in range(1,33):
    Separation2 = np.zeros((100,100))
    for i in range(100):
      # Separation.append([[0]*13]*13)
      Separation2[i][i]=1500
      for j in range(i+1, 100):
        Separation2[i][j]=abs(a.iloc[p*100+i]['PC{}'.format(k)]-a.iloc[p*100+j]['PC{}'.format(k)])
        Separation2[j][i]=Separation2[i][j]
    l1=[]
    for cls in range(100):
      l1.append(Separation2[cls][np.argmin(np.array(Separation2[cls]))])
      # l[cls][p*32+k-1] = l1[cls][k-1]
    l2[p][k-1] = np.mean(l1)
    l3[p][k-1] = np.median(l1)
```

```python
l2= pd.DataFrame(l2)
l2.describe()
```
