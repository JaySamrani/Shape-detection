## Standard Scaler

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scalar = StandardScaler().fit(X_train)
scalar
X_train= scalar.transform(X_train)
```

```python
X_test = scalar.transform(X_test)
X_test= pd.DataFrame(X_test)
```

```python
X_train= pd.DataFrame(X_train)
X_train.describe()
```

## PCA

```python
from sklearn.decomposition import PCA
# pca = PCA(n_components=5)
pca = PCA()
principalComponents = pca.fit_transform(X_train)
principalDf = pd.DataFrame(data = principalComponents, columns = ['PC{}'.format(i) for i in range(1,33,1)])
```

```python
principalDf
```

```python
pca.explained_variance_ratio_.shape, pca.explained_variance_ratio_
```

```python
Principal_Components = pca.components_
```

```python
PC = pd.DataFrame(Principal_Components)
```

```python
PC
```

```python
pca.get_params(deep=True)
```

```python
X_trainPCA = principalDf
```

```python
X_trainPCA
```

```python
pc = pca.transform(X_test)
X_testPCA = pd.DataFrame(data = pc, columns = ['PC{}'.format(i) for i in range(1,33,1)])
```

```python
X_testPCA
```

```python
X_trainPCA=X_trainPCA[['PC1','PC2','PC3','PC4','PC5','PC6','PC7','PC8']]
X_testPCA=X_testPCA[['PC1','PC2','PC3','PC4','PC5','PC6','PC7','PC8']]
```

## KNN

k = sqrt(n)/2 , n= no. of training data points

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=149)
history = neigh.fit(X_trainPCA, y_train)
```

```python
y_ = neigh.predict(X_testPCA)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_)
```

```python
neigh.get_params()
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test, y_)
```

```python
cm= pd.DataFrame(cm)
cm
```

```python
import seaborn as sb

dataplot = sb.heatmap(cm, cmap="YlGnBu", annot=True, annot_kws={"size": 5})

plt.show()
```

```python
for i in range(13):
  a=cm.iloc[i].sum()
  cm.iloc[i]= (cm.iloc[i]/a)*100

import seaborn as sb

dataplot = sb.heatmap(cm, cmap="YlGnBu", annot=True, annot_kws={"size": 5})

plt.show()
```

```python
from sklearn.metrics import precision_recall_fscore_support
print(precision_recall_fscore_support(y_test, y_, average='macro'),"\n")

print(precision_recall_fscore_support(y_test, y_, average='micro'),"\n")

print(precision_recall_fscore_support(y_test, y_, average='weighted'), "\n")
```
